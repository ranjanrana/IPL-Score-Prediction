```python
#removing unwanted collums/ cleaning the data
remove_data = ["mid" , "venue", "batsman","bowler","striker","non-striker"]
df.drop(labels=remove_data, axis=1, inplace=True)
```

```python
df.head()
```

```python
df["bat_team"].unique()
```

```python
actuve_team = ['Kolkata Knight Riders', 'Chennai Super Kings', 'Rajasthan Royals',
       'Mumbai Indians','Kings XI Punjab',
       'Royal Challengers Bangalore', 'Delhi Daredevils',
       'Sunrisers Hyderabad']
```

```python
df = df[(df['bat_team'].isin(actuve_team)) & (df['bowl_team'].isin(actuve_team))]
```

```python
# Removing the first powerplay data in every match
df = df[df['overs']>=5.0]
```

```python
df.head()
```

```python
print(df["bat_team"].unique())
print(df["bowl_team"].unique())
```

```python
# Converting the column 'date' from string into datetime object
from datetime import datetime
df['date'] = df['date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))
```

```python
df.head()
```

```python
#Data Preprocessing 
# Converting categorical features using OneHotEncoding method
method = pd.get_dummies(data=df, columns=['bat_team', 'bowl_team'])
```

```python
method.head()
```

```python
method.columns
```

```python
#reaarange the columns
method = method[['date','bat_team_Chennai Super Kings', 'bat_team_Delhi Daredevils',
       'bat_team_Kings XI Punjab', 'bat_team_Kolkata Knight Riders',
       'bat_team_Mumbai Indians', 'bat_team_Rajasthan Royals',
       'bat_team_Royal Challengers Bangalore', 'bat_team_Sunrisers Hyderabad',
       'bowl_team_Chennai Super Kings', 'bowl_team_Delhi Daredevils',
       'bowl_team_Kings XI Punjab', 'bowl_team_Kolkata Knight Riders',
       'bowl_team_Mumbai Indians', 'bowl_team_Rajasthan Royals',
       'bowl_team_Royal Challengers Bangalore',
       'bowl_team_Sunrisers Hyderabad', 'runs', 'wickets', 'overs', 'runs_last_5', 'wickets_last_5',
       'total']]
```

```python
method.head()
```

```python
# Splitting the data into train and test set
X_train = method.drop(labels='total', axis=1)[method['date'].dt.year <= 2016]
X_test = method.drop(labels='total', axis=1)[method['date'].dt.year >= 2017]
```

```python
y_train = method[method['date'].dt.year <= 2016]['total'].values
y_test = method[method['date'].dt.year >= 2017]['total'].values
```

```python
# Removing the 'date' column
X_train.drop(labels='date', axis=True, inplace=True)
X_test.drop(labels='date', axis=True, inplace=True)
```

```python
# Model Building
# Linear Regression Model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
regressor.intercept_
```

```python
regressor.coef_
```

```python
#Predict
y_pred = regressor.predict(X_test)
```

```python
#r2_score from metrics
from sklearn.metrics import r2_score
```

```python
values = r2_score(y_test,y_pred)
```
